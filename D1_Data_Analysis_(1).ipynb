## CLO1: Introduction and Data Wrangling

### Purpose of Data Analysis
The purpose of this analysis is to investigate the factors that influence the tip amount left by restaurant customers. By analyzing variables such as total bill, party size, gender, and time of visit, we aim to uncover patterns that can help restaurant management understand customer tipping behavior and improve service strategies.

### Type of Programming
Python is used for this analysis because it offers powerful libraries — Pandas for data wrangling, NumPy for numerical operations, Matplotlib and Seaborn for visualization, and SciPy for statistical testing. Its interactive environment via Jupyter Notebook makes it ideal for step-by-step exploratory data analysis.

### Type and Purpose of ML Algorithm
The ML algorithm to be implemented is **Linear Regression** (supervised learning). Its purpose is to predict the dependent variable `tip` based on numerical independent variables such as `total_bill` and `size`. It is appropriate because `tip` is a continuous numerical variable and a linear relationship is expected with bill amount.

### Identify the Variables
- **Dependent Variable:** `tip` — the tip amount in dollars (continuous, numerical)
- **Independent Variables:**
  - `total_bill` (Numerical) — the total restaurant bill in dollars
  - `size` (Numerical) — number of people in the dining party
  - `sex` (Categorical) — gender of the bill payer (Male/Female)
  - `time` (Categorical) — Lunch or Dinner

### Sampling
Yes, sampling will be performed. **Random sampling** (n=150) is applied first, giving every record an equal probability of selection — ensuring an unbiased subset. **Systematic sampling** is then applied by selecting every k-th record from customers who paid bills above \$10 (active spenders), providing a structured and efficient sample. Both methods are justified as they are representative and computationally efficient.

### CELL 1 — Load Dataset
> **📸 SCREENSHOT this cell's output → paste into Word doc under: CLO2 › Script for Descriptive Analysis**

```python
# ============================================================
# CELL 1: IMPORT LIBRARIES AND LOAD DATASET
# ============================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr, spearmanr, chi2_contingency, ttest_1samp

np.random.seed(42)

# Load the built-in Tips dataset from seaborn
df = sns.load_dataset('tips')

print('Dataset loaded successfully!')
print(f'Shape: {df.shape}')
print('\nFirst 10 rows:')
print(df.head(10))
print('\nData Types:')
print(df.dtypes)
print('\nMissing Values:')
print(df.isnull().sum())
```

---
### CELL 2 — Custom Descriptive Statistics Function
> **📸 SCREENSHOT this cell's output → paste into Word doc under: CLO2 › Script for Descriptive Analysis**

```python
# ============================================================
# CELL 2: CUSTOM DESCRIPTIVE STATISTICS FUNCTION
# ============================================================

def descriptive_stats(sample, field):
    """
    Custom function to compute descriptive statistics.

    Parameters:
        sample (DataFrame): The data sample
        field  (str)       : The column name to analyze

    Returns:
        dict: Dictionary containing all descriptive statistics
    """
    col = sample[field].dropna()

    result = {
        'Field'      : field,
        'Count'      : len(col),
        'Mean'       : round(col.mean(), 4),
        'Median'     : round(col.median(), 4),
        'Mode'       : round(col.mode()[0], 4),
        'Std Dev'    : round(col.std(), 4),
        'Variance'   : round(col.var(), 4),
        'Min'        : round(col.min(), 4),
        'Max'        : round(col.max(), 4),
        'Range'      : round(col.max() - col.min(), 4),
        'Q1 (25th)'  : round(col.quantile(0.25), 4),
        'Q3 (75th)'  : round(col.quantile(0.75), 4),
        'IQR'        : round(col.quantile(0.75) - col.quantile(0.25), 4),
        'Skewness'   : round(col.skew(), 4),
        'Kurtosis'   : round(col.kurtosis(), 4)
    }

    print(f"\n{'='*48}")
    print(f"  Descriptive Statistics for: [{field}]")
    print(f"{'='*48}")
    for key, val in result.items():
        print(f"  {key:<18}: {val}")
    print(f"{'='*48}")

    return result

# Apply function on full dataset
full_result = descriptive_stats(df, 'tip')
```

---
### CELL 3 — Random Sampling (size = 150)
> **📸 SCREENSHOT this cell's output → paste into Word doc under: CLO2 › Random Sampling**

```python
# ============================================================
# CELL 3: RANDOM SAMPLING  (n = 150, with replacement)
# ============================================================

# Random sample of size 150 (use replace=True since dataset has 244 rows)
random_sample = df.sample(n=150, replace=True, random_state=42)

print(f'Total Dataset Size : {len(df)}')
print(f'Random Sample Size : {len(random_sample)}')
print('\nRandom Sample Preview (first 10 rows):')
print(random_sample.head(10))

# Apply custom descriptive_stats function on random sample
print('\n--- Descriptive Statistics on Random Sample ---')
random_result = descriptive_stats(random_sample, 'tip')
```

---
### CELL 4 — Systematic Sampling
> **📸 SCREENSHOT this cell's output → paste into Word doc under: CLO2 › Systematic Sampling**

```python
# ============================================================
# CELL 4: SYSTEMATIC SAMPLING
# Condition: total_bill > 10  (customers with meaningful spend)
# ============================================================

# Step 1 – Apply condition
filtered_df = df[df['total_bill'] > 10].reset_index(drop=True)

# Step 2 – Calculate step size k
desired_size = 50
k = len(filtered_df) // desired_size

# Step 3 – Random start, then every k-th row
start = np.random.randint(0, k)
systematic_sample = filtered_df.iloc[start::k].head(desired_size)

print(f'Condition Applied          : total_bill > 10')
print(f'Filtered Population Size   : {len(filtered_df)}')
print(f'Step size (k)              : {k}')
print(f'Random Start Index         : {start}')
print(f'Systematic Sample Size     : {len(systematic_sample)}')
print('\nSystematic Sample Preview (first 10 rows):')
print(systematic_sample.head(10))

# Apply custom descriptive_stats function on systematic sample
print('\n--- Descriptive Statistics on Systematic Sample ---')
sys_result = descriptive_stats(systematic_sample, 'tip')
```

### CELL 5 — Descriptive Statistical Report
> **📸 SCREENSHOT this full output → paste into Word doc under: CLO2 › Descriptive Statistical Report**

```python
# ============================================================
# CELL 5: FULL DESCRIPTIVE STATISTICS REPORT
# ============================================================

print('=' * 55)
print('     DESCRIPTIVE STATISTICS REPORT — TIP AMOUNT')
print('=' * 55)

print('\n[1] Full Summary (.describe()):')
print(df['tip'].describe().round(4))

print('\n[2] Additional Statistics:')
print(f'  Skewness  : {df["tip"].skew():.4f}')
print(f'  Kurtosis  : {df["tip"].kurtosis():.4f}')
iqr = df['tip'].quantile(0.75) - df['tip'].quantile(0.25)
print(f'  IQR       : {iqr:.4f}')

print(f'\n[3] Missing Values: {df["tip"].isnull().sum()}')

print('\n[4] Tip Category Distribution:')
bins   = [0, 2, 4, 6, 10]
labels = ['Low ($0-2)', 'Medium ($2-4)', 'High ($4-6)', 'Very High (>$6)']
df['TipCategory'] = pd.cut(df['tip'], bins=bins, labels=labels, include_lowest=True)
print(df['TipCategory'].value_counts().sort_index())

print('\n[5] Average Tip by Gender:')
print(df.groupby('sex')['tip'].mean().round(4))

print('\n[6] Average Tip by Time (Lunch vs Dinner):')
print(df.groupby('time')['tip'].mean().round(4))
```

### CELL 7a — Pearson Correlation (Numerical IV: total_bill vs tip)
> **📸 SCREENSHOT output → paste into Word doc under: CLO2 › Hypothesis Test**

```python
# ============================================================
# CELL 7a: PEARSON CORRELATION
# IV: total_bill (numerical)  |  DV: tip (numerical)
# ============================================================

x = df['total_bill']
y = df['tip']

r_p, p_p = pearsonr(x, y)
df_p = len(x) - 2

print('=' * 55)
print('  PEARSON CORRELATION TEST')
print('=' * 55)
print('  H0: No significant linear correlation between total_bill and tip')
print('  H1: A significant linear correlation exists')
print(f'  Pearson r              : {r_p:.4f}')
print(f'  P-Value                : {p_p:.4f}')
print(f'  Degrees of Freedom     : {df_p}')

if p_p < 0.05:
    print(f'\n  Conclusion: p-value ({p_p:.4f}) < 0.05  -->  REJECT H0')
    print('  There IS a significant linear correlation between total_bill and tip.')
else:
    print(f'\n  Conclusion: p-value ({p_p:.4f}) >= 0.05  -->  ACCEPT H0')
    print('  There is NO significant linear correlation.')
```

---
### CELL 7b — Spearman Rank Correlation (total_bill vs tip)
> **📸 SCREENSHOT output → paste into Word doc under: CLO2 › Hypothesis Test**

```python
# ============================================================
# CELL 7b: SPEARMAN RANK CORRELATION
# IV: total_bill (numerical)  |  DV: tip (numerical)
# ============================================================

r_s, p_s = spearmanr(df['total_bill'], df['tip'])
df_s = len(df) - 2

print('=' * 55)
print('  SPEARMAN RANK CORRELATION TEST')
print('=' * 55)
print('  H0: No significant rank correlation between total_bill and tip')
print('  H1: A significant rank correlation exists')
print(f'  Spearman r             : {r_s:.4f}')
print(f'  P-Value                : {p_s:.4f}')
print(f'  Degrees of Freedom     : {df_s}')

if p_s < 0.05:
    print(f'\n  Conclusion: p-value ({p_s:.4f}) < 0.05  -->  REJECT H0')
    print('  There IS a significant rank correlation between total_bill and tip.')
else:
    print(f'\n  Conclusion: p-value ({p_s:.4f}) >= 0.05  -->  ACCEPT H0')
    print('  There is NO significant rank correlation.')
```

### CELL 7c — Chi-Square Test (Categorical IV: sex vs TipCategory)
> **📸 SCREENSHOT output → paste into Word doc under: CLO2 › Hypothesis Test**

```python
# ============================================================
# CELL 7c: CHI-SQUARE TEST
# IV: sex (categorical)  |  DV: TipCategory (derived from tip)
# ============================================================

# Make sure TipCategory column exists (created in Cell 5)
bins   = [0, 2, 4, 6, 10]
labels = ['Low', 'Medium', 'High', 'Very High']
df['TipCategory'] = pd.cut(df['tip'], bins=bins, labels=labels, include_lowest=True)

contingency = pd.crosstab(df['sex'], df['TipCategory'])
print('Contingency Table (sex vs TipCategory):')
print(contingency)

chi2, p_c, dof, expected = chi2_contingency(contingency)

print('\n' + '=' * 55)
print('  CHI-SQUARE TEST')
print('=' * 55)
print('  H0: No significant association between sex and TipCategory')
print('  H1: A significant association exists')
print(f'  Chi-Square Statistic   : {chi2:.4f}')
print(f'  P-Value                : {p_c:.4f}')
print(f'  Degrees of Freedom     : {dof}')

if p_c < 0.05:
    print(f'\n  Conclusion: p-value ({p_c:.4f}) < 0.05  -->  REJECT H0')
    print('  Sex and TipCategory are significantly associated.')
else:
    print(f'\n  Conclusion: p-value ({p_c:.4f}) >= 0.05  -->  ACCEPT H0')
    print('  No significant association between sex and TipCategory.')
```

### CELL 8 — One-Sample T-Test
> **📸 SCREENSHOT this output → paste into Word doc under: CLO2 › One Sample T-Test**

```python
# ============================================================
# CELL 8: ONE-SAMPLE T-TEST
# Assess if random sample tip mean = assumed population mean ($3.00)
# ============================================================

# Assumed population mean (industry benchmark: average tip = $3.00)
population_mean = 3.00

sample_tips = random_sample['tip']

t_stat, p_val = ttest_1samp(sample_tips, popmean=population_mean)
dof_t = len(sample_tips) - 1

print('=' * 55)
print('  ONE-SAMPLE T-TEST')
print('=' * 55)
print(f'  H0: Sample mean tip = population mean tip (mu = ${population_mean})')
print(f'  H1: Sample mean tip != population mean tip')
print(f'  Sample Size            : {len(sample_tips)}')
print(f'  Sample Mean            : ${sample_tips.mean():.4f}')
print(f'  Population Mean        : ${population_mean:.2f}')
print(f'  T-Statistic            : {t_stat:.4f}')
print(f'  P-Value                : {p_val:.4f}')
print(f'  Degrees of Freedom     : {dof_t}')
print(f'  Significance Level     : 0.05')

if p_val < 0.05:
    print(f'\n  Conclusion: p-value ({p_val:.4f}) < 0.05  -->  REJECT H0')
    print(f'  The sample mean significantly differs from ${population_mean}.')
    print('  The sample is NOT representative of this assumed population mean.')
else:
    print(f'\n  Conclusion: p-value ({p_val:.4f}) >= 0.05  -->  ACCEPT H0')
    print(f'  No significant difference from the assumed population mean of ${population_mean}.')
    print('  The sample IS representative of the population.')
```
